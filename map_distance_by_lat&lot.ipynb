```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv(r'D:\final project\Dataset\complete_orders.csv')

def haversine(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance in kilometers between two points 
    on the earth (specified in decimal degrees).
    """
    # Convert decimal degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    # Haversine formula
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a)) 
    return 6371.01 * c  # Radius of Earth in kilometers

def calculate_distances(df):
    """
    Calculate distances between customers and sellers in the DataFrame.
    Adds a new column 'distance_km' with calculated distances.
    """
    df['distance_km'] = haversine(df['geolocation_lat_customer'], 
                                   df['geolocation_lng_customer'], 
                                   df['geolocation_lat_seller'], 
                                   df['geolocation_lng_seller'])
    return df

# Calculate distances and add to DataFrame
df_with_distances = calculate_distances(df)

# Display the first few rows of the updated DataFrame with distances
print(df_with_distances[['geolocation_lat_customer', 'geolocation_lng_customer', 
                         'geolocation_lat_seller', 'geolocation_lng_seller', 
                         'distance_km']].head())

# If you want to save this updated DataFrame to a new CSV file
df_with_distances.to_csv(r'D:\final project\Dataset\complete_orders_with_distances.csv', index=False)
```
